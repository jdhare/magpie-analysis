```python
%matplotlib notebook
from scipy.special import erfi
import numpy as np
import matplotlib.pyplot as plt
```

```python
def B_Loureiro(x,d_sp):
    x0=0.92*d_sp
    if abs(x)<x0:
        return 1.85/2.0*np.sqrt(np.pi)*np.exp(-x**2/d_sp**2)*erfi(x/d_sp)
    if x>x0:
        return 1
    if x<-x0:
        return -1
```

```python
Bv=np.vectorize(B_Loureiro)
```
