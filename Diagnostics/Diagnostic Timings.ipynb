```python
filename="Carbon Reconnection Shot Book.csv"
shots=[]

with open(filename, 'r') as f:
    reader = csv.reader(f, delimiter=',')
    next(reader)
    for row in reader:
        shots.append(row[0])
        
shots=['s0426_15' if s=='s0426_16' else s for s in shots]
```

```python
ScopeChannel(shots[1], scope, channel)
```

# Ekspla Diode

```python
scope="2"
channel='A1'

fig, ax=plt.subplots(figsize=(12,4), sharex=True)
    
i=40
s=shots[i]
Ek=ScopeChannel(s, "2", "A1")
Fa=ScopeChannel(s, "2", "A2")
R1=ScopeChannel(s, "2", "C1")
R2=ScopeChannel(s, "2", "C2")

ax.plot(Ek.time,Ek.data)
ax.plot(Fa.time,Fa.data)
ax.plot(R1.time,R1.data)
ax.plot(R1.time,R2.data)

ax.set_title(s)
ax.set_xlim([1440,2000])

#banks


fig2, ax2=plt.subplots(figsize=(12,4), sharex=True)
for lg in ["A1", "A2", "B1", "B2"]:
    c=ScopeChannel(s, "10", lg)
    ax2.plot(c.time,c.data)

ax2.set_title(s)
ax2.set_xlim([1000,3000])
```
