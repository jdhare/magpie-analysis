# Rocket model for Density, Advection model for Magnetic Field

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as c
```

```python
#set up a current profile that is 0 before t=0 and 0 after 2*tau
def I_sin(t, tau, I0):
    if t<0:
        return 0
    if t>2*tau:
        return 0
    else:
        return I0*(np.sin(np.pi*t/(2*tau)))**2
I = np.vectorize(I_sin, otypes=[np.float])
```

## Parameters for model

```python
I0=1e6 #amps
tau=250e-9 #seconds
t=600e-9 #seconds
A=12 #atomic mass units
Z=4 #effective ion charge
V=100e3 #in m/s
R0=8e-3 #radius of array
lamb=1053e-9#laser probing wavelength for Faraday, metres.

xs=np.linspace(-80e-3,80e-3,1e3)# in metres, form is (minimum, maximum, number of points)
ys=np.linspace(9e-3,20e-3,1e3)# in metres, form is (minimum, maximum, number of points)
dx=(xs.max()-xs.min())/float(xs.size)
dy=(ys.max()-ys.min())/float(ys.size)
X, Y= np.meshgrid(xs, ys)
R=np.sqrt(Y**2+X**2) #distance from centre of array to each point, m
ret_time=t-(R-R0)/V#retarded time taken to reach each point, seconds
I_ret=I(ret_time, tau, I0)
ne=(c.mu_0*Z)/(8.0*np.pi**2*V**2*R0*R*A*c.m_p)*I_ret**2 #electron density, m^-3
B=c.mu_0/(2*np.pi*R0)*I_ret#magnetic field magnitude, T
th=np.arctan(X/Y)#angle between x axis and line from the radius
b_hat=np.array([[np.cos(the),np.sin(the)] for the in th]) #unit vector pointing in mangetic field direction
Bx=B*b_hat[:,0] #component of B field in x direction

neL=1e-4*np.sum(ne, axis=1)*dx #line average electron density cm^-2
alpha=2.62e-13*lamb**2*np.sum(Bx*ne, axis=1)*dx #rotation angle, radians
Bmax=B[:, ys.size/2] #peak magnetic field along any radial direction - this is what we are trying to extract
Bavg=3.4e20*alpha/(neL) #x component of magnetic field inferred by current analysis techniques.
```

## Single Exploder Model

### Plot profiles along chord of n_e and B

```python
y=15e-3#metres, distance from centre in direction perpendicular to line of sight
y_ind=np.where(y<ys)[0][0]

fig, ax=plt.subplots(figsize=(16,10))
#ax.plot(xs*1e3, B[y_ind,:], lw=5, color='k', label=r'B in tesla') #convert to degrees
ax.plot(xs*1e3, Bx[y_ind,:], lw=5, color='gray', label=r'B in tesla') #convert to degrees

ax.plot(xs*1e3, ne[y_ind,:]/1e23,lw=5,color='b', label=r'$n_e$ in $\times 10^{17}\,cm^{-3}$')
ax.legend(fontsize=20)

ax.set_xlabel(r'Distance along chord (mm)',fontsize=20)
ax.set_ylabel('Density/Tesla',fontsize=20)
ax.tick_params(labelsize=20, pad=5, length=10, width=2)

I0str=r'$I_0=$'+str(I0/1e6)+' MA, '
tstr=r'$t=$'+str(t*1e9)+' ns, '
Vstr=r'$V=$'+str(V/1e3)+' km/s, '
ax.set_title('Model for '+I0str+tstr+Vstr, fontsize=30)

ax.set_xlim([-80,80])
```

## Double Exploder Model

### Plot profiles perpendicular to line of sight with interaction region

```python
def interaction_average(ys, quantity, R_sym, d_int):
    dy=(ys.max()-ys.min())/float(ys.size)
    cutoff_ind=np.where((R_sym-d_int)<ys)[0][0]
    sym_ind=np.where(R_sym<ys)[0][0]
    quantity=np.nan_to_num(quantity)
    interaction_integrated_averaged=np.sum(quantity[cutoff_ind:])*dy/(d_int)
    stitched_array=np.zeros_like(quantity)
    stitched_array[:cutoff_ind]=quantity[:cutoff_ind]
    stitched_array[cutoff_ind:sym_ind]=interaction_integrated_averaged
    return stitched_array

R_sym=13.5e-3#distance to plane of symmetry
d_int=1.2e-3#half width of interaction region

neL_int=interaction_average(ys, neL, R_sym, d_int)
Bx_int=interaction_average(ys, Bavg, R_sym, d_int)

fig, ax=plt.subplots(figsize=(16,10))
ax.plot(ys*1e3, neL_int/1e17, lw=5, color='b', label=r'Interacting $\langle n_e L \rangle$ in $\times 10^{17}\,cm^{-2}$')
ax.plot(ys*1e3, neL/1e17, lw=5, ls=':', color='b', label=r'Non-interacting $\langle n_e L \rangle$ in $\times 10^{17}\,cm^{-2}$')
ax.plot(ys*1e3, Bx_int, lw=5, color='k', label=r'Interacting $B_{avg}$ in T')
ax.plot(ys*1e3, Bavg, lw=5, ls=':',color='k', label=r'Non-interacting $B_{avg}$ in T')
ax.legend(fontsize=20)

ax.set_xlabel(r'Distance from array centre (mm)',fontsize=20)
ax.set_ylabel('Density/Tesla',fontsize=20)
ax.tick_params(labelsize=20, pad=5, length=10, width=2)

I0str=r'$I_0=$'+str(I0/1e6)+' MA, '
tstr=r'$t=$'+str(t*1e9)+' ns, '
Vstr=r'$V=$'+str(V/1e3)+' km/s, '
R_sym_str=r'$R_{sym}=$'+str(R_sym*1e3)+' mm, '
d_int_str=r'$d_{int}=$'+str(d_int*1e3)+' mm, '
ax.set_title('Model for '+I0str+tstr+Vstr+' and '+R_sym_str+d_int_str, fontsize=23)
```

# Time Dependent Model

```python
I0=1e6 #amps
tau=250e-9 #seconds
A=12 #atomic mass units
Z=4 #effective ion charge
V=100e3 #in m/s
R0=8e-3 #radius of array
lamb=1550e-9#laser probing wavelength for Faraday, metres.

x0=20e-3
ys=np.linspace(-40e-3,40e-3,1e3)# in metres, form is (minimum, maximum, number of points)
R=np.hypot(x0,ys)
th=np.arctan(ys/x0)#angle between x axis and line from the radius
b_y=np.array([np.cos(the) for the in th]) #unit vector pointing in mangetic field direction

t=np.linspace(0, 1000e-9, 1000)
neL=np.zeros_like(t)
alpha=np.zeros_like(t)
Bmax=np.zeros_like(t)
Bavg=np.zeros_like(t)

for i,tt in enumerate(t):
    ret_time=tt-(R-R0)/V#retarded time taken to reach each point, seconds
    I_ret=I(ret_time, tau, I0)
    ne=(c.mu_0*Z)/(8.0*np.pi**2*V**2*R0*R*A*c.m_p)*I_ret**2 #electron density, m^-3
    B=c.mu_0/(2*np.pi*R0)*I_ret#magnetic field magnitude, T
    By=B*b_y #component of B field in x direction

    neL[i]=1e-4*np.sum(ne)*dx #line average electron density cm^-2
    alpha[i]=2.62e-13*lamb**2*np.sum(By*ne)*dx #rotation angle, radians
    Bmax[i]=By.max() #peak magnetic field along any radial direction - this is what we are trying to extract
    Bavg[i]=3.82e12/(lamb**2)*alpha[i]/(1e4*neL[i]) #x component of magnetic field inferred by current analysis techniques.
```

```python
fig, ax1=plt.subplots(figsize=(6,4))
ax1.plot(t*1e9, neL/1e17, lw=2, color='k') 
ax1.set_xlabel(r'Time (ns)',fontsize=10)
ax1.set_ylabel(r'$\langle n_eL \rangle$ ($\times 10^{17}\,cm^{-2}$)',fontsize=10)
ax1.tick_params(labelsize=10, pad=5, length=10, width=2)
#ax1.plot(t*1e9, Bmax, lw=2, color='r') 
#ax1.plot(t*1e9, Bavg, lw=2, color='g') 


ax2 = ax1.twinx()
ax2.plot(t*1e9, alpha,lw=2,color='b')
ax2.set_ylabel(r'$\alpha$ (radians)',fontsize=10, color='b')
ax2.tick_params(labelsize=10, pad=5, length=10, width=2, colors='b')


I0str=r'$I_0=$'+str(I0/1e6)+' MA, '
xstr=r'$x_0=$'+str(x0*1e3)+' mm, '
Vstr=r'$V=$'+str(V/1e3)+' km/s, '
fig.suptitle('Model for '+I0str+xstr+Vstr, fontsize=12)

#ax.set_xlim([-10,10])
```
